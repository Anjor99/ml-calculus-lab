Imports

```python
import sys
import os
sys.path.append(os.path.abspath(".."))

from calculus.gradients import gradient_descent_1d
```

Define the function and its derivative

```python
def f(x):
    return x ** 2

def df(x):
    return 2 * x
```

Call gradient descent

```python
x_min, f_min, history = gradient_descent_1d(
    f,
    df,
    x0=5.0,
    learning_rate=0.1
)

print("Minimum x:", x_min)
print("Minimum f(x):", f_min)
```

Visualize the results

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(history, marker='o')
plt.xlabel("Iteration")
plt.ylabel("x value")
plt.title("Gradient Descent Convergence (x)")
plt.show()
```
