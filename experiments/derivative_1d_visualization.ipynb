1️⃣ Imports (Top Cell)

```python
import sys
import os
sys.path.append(os.path.abspath(".."))

import numpy as np
import matplotlib.pyplot as plt

from calculus.derivatives import numerical_derivative
```

2️⃣ Define the Function

```python
def f(x):
    return x**2
```

3️⃣ Choose a Point to Differentiate

```python
x0 = 1.0
slope = numerical_derivative(f, x0)

print("Slope at x =", x0, "is", slope)
```

4️⃣ Create x-values for Plotting

```python
x = np.linspace(-3, 3, 400)
y = f(x)
```

5️⃣ Equation of the Tangent Line

```python
x_tan = np.linspace(x0 - 0.8, x0 + 0.8, 100)
y_tan = f(x0) + slope * (x_tan - x0)
```

6️⃣ Plot Everything

```python
plt.figure()
plt.plot(x, y, label="f(x) = x²")
plt.plot(x_tan, y_tan, linestyle="--", linewidth=2, label="Tangent at x = 1")
plt.scatter([x0], [f(x0)])
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
